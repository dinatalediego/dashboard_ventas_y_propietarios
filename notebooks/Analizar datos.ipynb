## Contarme como fechas

```python
import pandas as pd


def describir_fechas(df, texto_busqueda="fecha", dayfirst=True):
    """
    Encuentra columnas relacionadas con fechas y devuelve:
    - tipo original;
    - cantidad de registros;
    - fechas válidas e inválidas;
    - fecha mínima y máxima;
    - amplitud temporal en días.

    No modifica el DataFrame original.
    """

    # Busca columnas que contengan "fecha"
    columnas_fecha = [
        columna
        for columna in df.columns
        if texto_busqueda.lower() in columna.lower()
        or pd.api.types.is_datetime64_any_dtype(df[columna])
    ]

    if not columnas_fecha:
        print(
            f"No encontré columnas que contengan '{texto_busqueda}' "
            "ni columnas reconocidas como datetime."
        )
        return pd.DataFrame()

    resumen = []

    for columna in columnas_fecha:

        serie_original = df[columna]

        # Intenta convertir los valores a fecha
        serie_fecha = pd.to_datetime(
            serie_original,
            errors="coerce",
            dayfirst=dayfirst
        )

        registros_con_dato = serie_original.notna().sum()
        fechas_validas = serie_fecha.notna().sum()
        fechas_invalidas = registros_con_dato - fechas_validas

        fecha_minima = serie_fecha.min()
        fecha_maxima = serie_fecha.max()

        if pd.notna(fecha_minima) and pd.notna(fecha_maxima):
            rango_dias = (fecha_maxima - fecha_minima).days
        else:
            rango_dias = None

        resumen.append({
            "columna": columna,
            "tipo_original": str(serie_original.dtype),
            "total_filas": len(df),
            "registros_con_dato": registros_con_dato,
            "fechas_validas": fechas_validas,
            "fechas_invalidas": fechas_invalidas,
            "valores_nulos": serie_original.isna().sum(),
            "fecha_inicio": fecha_minima,
            "fecha_final": fecha_maxima,
            "rango_dias": rango_dias
        })

    resultado = pd.DataFrame(resumen)

    return resultado.sort_values(
        by="fecha_inicio",
        na_position="last"
    ).reset_index(drop=True)

resumen_fechas = describir_fechas(df)

resumen_fechas
```

```python
def contar_historia_fechas(df):
    resumen = describir_fechas(df)

    if resumen.empty:
        return

    print("Hola, estas son las columnas de fecha que encontré:\n")

    for _, fila in resumen.iterrows():

        columna = fila["columna"]
        inicio = fila["fecha_inicio"]
        final = fila["fecha_final"]
        validas = fila["fechas_validas"]
        invalidas = fila["fechas_invalidas"]
        nulos = fila["valores_nulos"]

        if pd.isna(inicio) or pd.isna(final):
            print(
                f"• {columna}: no pude encontrar un rango de fechas válido."
            )
            continue

        print(
            f"• En '{columna}', mi historia comienza el "
            f"{inicio:%d/%m/%Y} y termina el {final:%d/%m/%Y}."
        )

        print(
            f"  Reconocí {validas:,} fechas válidas, "
            f"{invalidas:,} valores inválidos y "
            f"{nulos:,} valores nulos.\n"
        )

contar_historia_fechas(df)
```

## gráfico de fechas

```python
import plotly.express as px


def graficar_rangos_fechas(df):
    resumen = describir_fechas(df)

    resumen = resumen.dropna(
        subset=["fecha_inicio", "fecha_final"]
    )

    if resumen.empty:
        print("No hay columnas con rangos de fechas válidos.")
        return

    figura = px.timeline(
        resumen,
        x_start="fecha_inicio",
        x_end="fecha_final",
        y="columna",
        title="Cobertura temporal de las columnas de fecha",
        hover_data={
            "fechas_validas": ":,",
            "fechas_invalidas": ":,",
            "valores_nulos": ":,",
            "rango_dias": ":,"
        }
    )

    figura.update_yaxes(
        title="Columnas",
        autorange="reversed"
    )

    figura.update_xaxes(
        title="Periodo cubierto",
        showgrid=True
    )

    figura.update_layout(
        template="plotly_white",
        title_x=0.02,
        height=max(400, len(resumen) * 65),
        margin=dict(l=40, r=40, t=80, b=40),
        hoverlabel=dict(
            bgcolor="white",
            font_size=13
        )
    )

    figura.show()
```

```python
graficar_rangos_fechas(df)
```

## Preparar distribucion de fechas

```python
def preparar_distribucion_fechas(
    df,
    frecuencia="M",
    texto_busqueda="fecha",
    dayfirst=True
):
    """
    frecuencia:
    D = día
    W = semana
    M = mes
    Q = trimestre
    Y = año
    """

    columnas_fecha = [
        columna
        for columna in df.columns
        if texto_busqueda.lower() in columna.lower()
        or pd.api.types.is_datetime64_any_dtype(df[columna])
    ]

    distribuciones = []

    for columna in columnas_fecha:

        fechas = pd.to_datetime(
            df[columna],
            errors="coerce",
            dayfirst=dayfirst
        ).dropna()

        if fechas.empty:
            continue

        distribucion = (
            fechas
            .dt.to_period(frecuencia)
            .value_counts()
            .sort_index()
            .rename_axis("periodo")
            .reset_index(name="cantidad")
        )

        distribucion["fecha"] = distribucion["periodo"].dt.start_time
        distribucion["columna"] = columna

        distribuciones.append(
            distribucion[["fecha", "columna", "cantidad"]]
        )

    if not distribuciones:
        return pd.DataFrame(
            columns=["fecha", "columna", "cantidad"]
        )

    return pd.concat(
        distribuciones,
        ignore_index=True
    )
preparar_distribucion_fechas(df)
```

### Perfil dataframe

```python
import pandas as pd
import numpy as np


def perfilar_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    """
    Genera un resumen de calidad y estructura para cada columna
    de un DataFrame.
    """

    filas = len(df)
    resultados = []

    for columna in df.columns:
        serie = df[columna]

        no_nulos = serie.notna().sum()
        nulos = serie.isna().sum()
        unicos = serie.nunique(dropna=True)

        porcentaje_nulos = (
            round(nulos / filas * 100, 2)
            if filas > 0
            else 0
        )

        porcentaje_unicos = (
            round(unicos / no_nulos * 100, 2)
            if no_nulos > 0
            else 0
        )

        ejemplo = serie.dropna().iloc[0] if no_nulos > 0 else None

        resultados.append({
            "variable": columna,
            "tipo_python": str(serie.dtype),
            "filas": filas,
            "no_nulos": no_nulos,
            "nulos": nulos,
            "porcentaje_nulos": porcentaje_nulos,
            "valores_unicos": unicos,
            "porcentaje_unicos": porcentaje_unicos,
            "ejemplo": ejemplo
        })

    perfil = pd.DataFrame(resultados)

    return perfil.sort_values(
        by=["porcentaje_nulos", "valores_unicos"],
        ascending=[False, True]
    ).reset_index(drop=True)


perfil = perfilar_dataframe(df)
with pd.option_context("display.max_rows", None):
    display(perfil)
#display(perfil)
```
